# **Exploratory Data Analysis**

```python
sns.set(style='dark')
plt.subplots(figsize=(10,5))
sns.regplot(x=data['Sales'],y=data['Profit'],color='green',line_kws={'color':'red'})
plt.xlabel("SALES",fontdict={'color':'black','fontsize':15})
plt.ylabel("PROFIT",fontdict={'color':'black','fontsize':15})
plt.title("RELATIONSHIP BETWEEN SALES AND PROFIT",fontdict={'color':'black','fontsize':20})
```

Here from the scatterplot we see that there is a positively sloped trendline which concludes a strong positive relationship between sales and profit.
